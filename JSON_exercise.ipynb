## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
df = pd.read_json('data/world_bank_projects.json')
df.head()
```

### Question 1: Top 10 countries with most project:

```python
# Isolating, counting, sorting in descending order and saving in unique variable.
df_most_projects = pd.DataFrame(df.groupby('countryname')['project_name'].count().sort_values(ascending=False))
df_most_projects.reset_index(inplace=True)
df_most_projects.columns = ['Country Name', 'Project Name']
df_most_projects.head(10)
```

### Question 2: Top 10 themes used in project:

```python
# Isolating the column in questions:
df_themes = df['mjtheme_namecode']
df_themes_normalized = pd.concat([pd.DataFrame(json_normalize(x)) for x in df_themes], ignore_index=True)
df_themes_normalized.head()
```

```python
# Sorting themes in descending order:
pd.DataFrame(df_themes_normalized['name'].value_counts())
```

### Question 3: Filling missing themes:

```python
# Tying codes with names:
df_themes_clean = df_themes_normalized.groupby('name')['code'].max()
# Selecting only codes with names:
df_themes_clean = pd.DataFrame(df_themes_clean[df_themes_clean.index != ''], columns=['code'])
# Setting the indexes to code:
df_themes_clean['clean_name'] = df_themes_clean.index
df_themes_clean = df_themes_clean.set_index(['code'])
print(df_themes_clean)
```

```python
# Merging df_themes_clean with df_themes_normalized to fill in the blank rows:
df_themes_normalized_clean = df_themes_normalized.merge(df_themes_clean, how='outer', left_on=['code'], right_index=True)
df_themes_normalized_clean.head()
```

```python
# Recounting the number of themes with filled rows:
df_themes_normalized_clean['clean_name'].value_counts()
```

```python
# sns.barplot(df_themes_normalized['clean_name'].value_counts())
# df_themes_normalized_clean = df_themes_normalized_clean.drop('name', axis=1)
plt.figure(figsize=(20, 7.5))

sns.countplot(df_themes_normalized_clean['clean_name'])
plt.title('Total count of each theme:', fontsize=20)
plt.xlabel('', fontsize=15)
plt.ylabel('')
plt.xticks(fontsize=15, rotation=90)
plt.yticks(fontsize=15)

plt.show()
```
